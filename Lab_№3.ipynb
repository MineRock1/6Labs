```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_half = cv2.imread("/content/Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread("/content/Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_half, cmap = 'gray');
plt.subplot(122)
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.imshow(img_chess, cmap = 'gray');
plt.subplot(122)
hist_chest = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
plt.plot(hist_chest); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.show();
```



---


**Застосуємо розмивання Гауса та звичайний згладжувальний фільтр до вхідного зображення**

```python
gblur_half3 = cv2.GaussianBlur(img_half,(3,3),0)
gblur_half5 = cv2.GaussianBlur(img_half,(5,5),0)
gblur_half9 = cv2.GaussianBlur(img_half,(9,9),0)
gblur_half15 = cv2.GaussianBlur(img_half,(15,15),0)

gblur_chess3 = cv2.GaussianBlur(img_chess,(3,3),0)
gblur_chess5 = cv2.GaussianBlur(img_chess,(5,5),0)
gblur_chess9 = cv2.GaussianBlur(img_chess,(9,9),0)
gblur_chess15 = cv2.GaussianBlur(img_chess,(15,15),0)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(521)
plt.imshow(img_half, cmap = 'gray')
plt.title('Original')
plt.subplot(522)
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(523)
plt.imshow(gblur_half3, cmap = 'gray')
plt.title('Gaussian Blur 3x3')
plt.subplot(524)
hist_half = cv2.calcHist([gblur_half3], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(525)
plt.imshow(gblur_half5, cmap = 'gray')
plt.title('Gaussian Blur 5x5')
plt.subplot(526)
hist_half = cv2.calcHist([gblur_half5], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(527)
plt.imshow(gblur_half9, cmap = 'gray')
plt.title('Gaussian Blur 9x9')
plt.subplot(528)
hist_half = cv2.calcHist([gblur_half9], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(529)
plt.imshow(gblur_half15, cmap = 'gray')
plt.title('Gaussian Blur 15x15')
plt.subplot(5,2,10)
hist_half = cv2.calcHist([gblur_half15], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show()




plt.figure(figsize=(20,20))
plt.subplot(521)
plt.imshow(img_chess, cmap = 'gray')
plt.title('Original')
plt.subplot(522)
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(523)
plt.imshow(gblur_chess3, cmap = 'gray')
plt.title('Gaussian Blur 3x3')
plt.subplot(524)
hist_chess = cv2.calcHist([gblur_chess3], [0], None, [256], [0, 256])
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(525)
plt.imshow(gblur_chess5, cmap = 'gray')
plt.title('Gaussian Blur 5x5')
plt.subplot(526)
hist_chess = cv2.calcHist([gblur_chess5], [0], None, [256], [0, 256])
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(527)
plt.imshow(gblur_chess9, cmap = 'gray')
plt.title('Gaussian Blur 9x9')
plt.subplot(528)
hist_chess = cv2.calcHist([gblur_chess9], [0], None, [256], [0, 256])
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show()

plt.figure(figsize=(20,20))
plt.subplot(529)
plt.imshow(gblur_chess15, cmap = 'gray')
plt.title('Gaussian Blur 15x15')
plt.subplot(5,2,10)
hist_chess = cv2.calcHist([gblur_chess15], [0], None, [256], [0, 256])
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show()
```
