```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bw = cv2.imread("/content/batman.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(5, 5))
plt.imshow(img_bw, cmap='gray')
plt.show()
```

**ÐŸÐµÑ€ÐµÑ‚Ð²Ð¾Ñ€ÐµÐ½Ð½Ñ BGR to RGB**

```python
img_joker = cv2.imread("/content/joker.jpg")

img_rgb = cv2.cvtColor(img_joker, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5, 5))
plt.imshow(img_joker)
plt.show()

plt.figure(figsize=(5, 5))
plt.imshow(img_rgb)
plt.show();
```

**ÐŸÐµÑ€ÐµÑ‚Ð²Ð¾Ñ€ÐµÐ½Ð½Ñ RGB to HSV**

```python
img_hsv = cv2.cvtColor(img_joker, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(5, 5))
plt.imshow(img_hsv)
plt.show();
```

**ÐžÐºÑ€ÐµÐ¼Ð¾ Ð²Ñ–Ð´Ð¾Ð±Ñ€Ð°Ð·Ð¸Ñ‚Ð¸ ÐºÐ¾Ð»Ñ–Ñ€Ð½Ñ– ÐºÐ°Ð½Ð°Ð»Ð¸ Ð´Ð»Ñ Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½ÑŒ Ñ–Ð· ÐºÐ¾Ð»Ñ–Ñ€Ð½Ð¸Ð¼Ð¸ Ð¼Ð¾Ð´ÐµÐ»ÑÐ¼Ð¸ RGB Ñ‚Ð° HSV**

```python
plt.figure(figsize=(20, 20))

plt.subplot(131);
R = np.zeros(img_rgb.shape, dtype='uint8')
R[:, :, 0] = img_rgb[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_rgb.shape, dtype='uint8')
G[:, :, 1] = img_rgb[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_rgb.shape, dtype='uint8')
B[:, :, 2] = img_rgb[:, :, 2]
plt.imshow(B)

plt.show()
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(131);
R = np.zeros(img_hsv.shape, dtype='uint8')
R[:, :, 0] = img_hsv[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_hsv.shape, dtype='uint8')
G[:, :, 1] = img_hsv[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_hsv.shape, dtype='uint8')
B[:, :, 2] = img_hsv[:, :, 2]
plt.imshow(B)

plt.show()
```
